```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Data Loading and Analysis ---

# Define the file path
file_path = 'bird_migration_dataset-1.xlsx'

# Load the dataset from the Excel file
df = pd.read_excel(file_path)

# Count the occurrences of each starting location
start_location_counts = df['Start_Location'].value_counts()

# Count the occurrences of each ending location
end_location_counts = df['End_Location'].value_counts()

# --- Visualization for Start Locations ---

plt.figure(figsize=(12, 7))
ax1 = start_location_counts.plot(kind='bar', color='darkblue')

# Set the title and labels for the Start Location chart
plt.title('Most Frequent Starting Locations', fontsize=16)
plt.xlabel('Start Location', fontsize=12)
plt.ylabel('Number of Migrations', fontsize=12)

# Add numerical values on top of each bar
for p in ax1.patches:
    ax1.text(p.get_x() + p.get_width() / 2,
             p.get_height(),
             f"{int(p.get_height())}",
             ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('most_frequent_start_locations.png')
plt.show()

print("Counts for Starting Locations:")
print(start_location_counts)

# --- Visualization for End Locations ---

plt.figure(figsize=(12, 7))
ax2 = end_location_counts.plot(kind='bar', color='darkred')

# Set the title and labels for the End Location chart
plt.title('Most Frequent Ending Locations', fontsize=16)
plt.xlabel('End Location', fontsize=12)
plt.ylabel('Number of Migrations', fontsize=12)

# Add numerical values on top of each bar
for p in ax2.patches:
    ax2.text(p.get_x() + p.get_width() / 2,
             p.get_height(),
             f"{int(p.get_height())}",
             ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('most_frequent_end_locations.png')
plt.show()

print("\nCounts for Ending Locations:")
print(end_location_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Data Loading and Analysis ---

# Define the file path
file_path = 'bird_migration_dataset-1.xlsx'

# Load the dataset from the Excel file
df = pd.read_excel(file_path)

# Calculate the average migration distance for each season
seasonal_avg_distance = df.groupby('Season')['Distance_km'].mean()

# Sort the values for better visualization
seasonal_avg_distance = seasonal_avg_distance.sort_values(ascending=False)

# --- Visualization ---

# Create the bar chart and assign it to a variable 'ax'
plt.figure(figsize=(10, 6))
ax = seasonal_avg_distance.plot(kind='bar', color='darkorange')

# Set the title and labels
plt.title('Average Migration Distance by Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Average Distance (km)', fontsize=12)

# Add numerical values on top of each bar
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2,
            p.get_height(),
            f"{p.get_height():.2f}",
            ha='center', va='bottom', fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Adjust layout to prevent labels from being cut off
plt.tight_layout()

# Save the plot to a file
plt.savefig('average_distance_by_season.png')

# Display the plot
plt.show()

# Print the data used for the chart
print("Average Migration Distance by Season:")
print(seasonal_avg_distance)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Data Loading and Analysis ---

# Define the file path
file_path = 'bird_migration_dataset-1.xlsx'

# Load the dataset from the Excel file
df = pd.read_excel(file_path)

# Create a new column for the migration route by combining start and end locations
df['Migration_Route'] = df['Start_Location'] + ' -> ' + df['End_Location']

# Count the occurrences of each unique migration route
route_counts = df['Migration_Route'].value_counts()

# --- Visualization ---

# Create the bar chart and assign it to a variable 'ax'
plt.figure(figsize=(12, 7))
ax = route_counts.plot(kind='bar', color='darkviolet')

# Set the title and labels
plt.title('Most Common Migration Routes', fontsize=16)
plt.xlabel('Migration Route', fontsize=12)
plt.ylabel('Number of Migrations', fontsize=12)

# Add numerical values on top of each bar
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2,
            p.get_height(),
            f"{int(p.get_height())}",
            ha='center', va='bottom', fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent labels from being cut off
plt.tight_layout()

# Save the plot to a file
plt.savefig('most_common_migration_routes.png')

# Display the plot
plt.show()

# Print the data used for the chart
print("Migration Route Counts:")
print(route_counts)
```
